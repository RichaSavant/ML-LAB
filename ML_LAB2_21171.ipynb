```python
import numpy as np
import pandas as py
from scipy import stats
from sklearn.preprocessing import StandardScaler

data = py.read_excel(r"C:\Amrita cse\5th Semester\Machine Learning\Lab\Lab 2\19CSE305_LabData_Set3.1.xlsx",sheet_name="thyroid0387_UCI")

df = py.DataFrame(data)

print(df.dtypes)
```

```python
data_types = {}
# Iterate over each column
for column in df.columns:
    # Get the unique values in the column
    unique_values = df[column].unique()
    # Determine if the datatype is ordinal or nominal
    if len(unique_values) > 10:
        data_types[column] = "Ordinal"
    else:
        data_types[column] = "Nominal"
# Print the datatype of each attribute
for column, data_type in data_types.items():
    print("Attribute:", column)
    print("Datatype:", data_type)
    print()
```

A3. Data Normalization / Scaling: from the data study, identify the attributes which may need
normalization. Employ appropriate normalization techniques to create normalized set of data.

Min-Max Scaling

```python
# Select the numeric columns you want to normalize
numeric_columns = df.select_dtypes(include=['number'])

# Define the minimum and maximum values for the scaling range
min_range = 0
max_range = 1

# Calculate the minimum and maximum values for each selected column
min_values = numeric_columns.min()
max_values = numeric_columns.max()

# Apply Min-Max scaling to the selected columns
min_max_scaled = (numeric_columns - min_values) / (max_values - min_values) * (max_range - min_range) + min_range

# Replace the original columns with the scaled values
df[numeric_columns.columns] = min_max_scaled

print("Scaled DataFrame:")
print(df)
```

Z-score Normalization

```python
# Select only the numeric columns you want to normalize
numeric_columns = df.select_dtypes(include=['number'])

# Apply Z-Score scaling using Scikit-Learn's StandardScaler
scaler = StandardScaler()
z_score_scaled = scaler.fit_transform(numeric_columns)

# Replace the original columns with the scaled values
df[numeric_columns.columns] = z_score_scaled
print("Scaled DataFrame:")
print(df)
```
